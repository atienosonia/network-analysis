```python
import pandas as pd
df = pd.read_csv('cleaned_data.csv')
df.info()
```

```python
# Extract unique authors with their attributes
nodes_df = df[['author', 'twitter_author_id', 'gender', 'city']].drop_duplicates().reset_index(drop=True)

# Add a unique numeric ID for each node
nodes_df['Id'] = range(1, len(nodes_df) + 1)

# Add a label column (same as Author for simplicity)
nodes_df['Label'] = nodes_df['author']

# Reorder columns for convenience
nodes_df = nodes_df[['Id', 'Label', 'twitter_author_id', 'gender', 'city']]

# Save the nodes DataFrame to CSV
nodes_df.to_csv('nodes.csv', index=False)
nodes_df
```

```python
import pandas as pd

# Initialize an empty list to store edges
edges = []

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    author = row['author']
    mentioned_authors = row['mentioned_authors']
    
    # Check if mentioned_authors is not null and not empty
    if pd.notna(mentioned_authors) and mentioned_authors:
        # Split mentioned_authors by comma and strip any leading/trailing spaces
        mentioned_list = mentioned_authors.split(',')
        
        # Create an edge for each mentioned author
        for mentioned_author in mentioned_list:
            edge = {
                'Source': author.strip(),
                'Target': mentioned_author.strip(),
                'Type': 'Mention'
            }
            edges.append(edge)

# Convert the list of edges to a DataFrame
edges_df = pd.DataFrame(edges)

# Drop duplicate edges based on Source, Target, and Type columns
edges_df = edges_df.drop_duplicates()

# Save the edges DataFrame to a CSV file
edges_df.to_csv('edges.csv', index=False)
edges_df
```
